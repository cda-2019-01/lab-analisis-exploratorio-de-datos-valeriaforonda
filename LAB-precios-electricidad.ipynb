Evolución de los Precios de la Electricidad
===

# Instrucciones

En el archivo `precios.zip`  se encuentran los archivos `Precio_Bolsa_Nacional_($kwh)_ano.xls` en formato de Microsoft Excel, los cuales contienen los precios históricos horarios de la electricidad para el mercado eléctrico Colombiano entre los años 1995 y 2017 en COL-PESOS/kWh. A partir de la información suministrada realice las siguientes actividades. 

**1.--** Lea los archivos y cree una tabla única concatenando la información para cada uno de los años. Debe transformar la tabla de tal forma que quede con las columnas `Fecha`, `Hora` y `Precio` (únicamente tres columnas).

A continuación se importan las liberías necesarias y la función para graficar online.
Se crea una función que para cada archivo busca en las primeras 10 filas de cada archivo hasta que se encuentre con la palabra "Fecha" que es la clave a partir de la cual va a considerar los datos con los que va a trabajar. 
A continuación, se crea una lista con los nombres de los archivos que están en la carpeta de datos, y se aplica la función previamente creada sobre cada uno de ellos. Todo esto se concatena en una única tabla. 

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import datetime as dt
import glob

%matplotlib inline

def Lectura(filename):
    for indice in range(10):
        DataFrame = pd.read_excel(filename,skiprows = indice ,usecols = list(range(25)))
        if DataFrame.iloc[0,0] == 'Fecha':
            DataFrame = pd.read_excel(filename,skiprows = indice+1, usecols = list(range(25)))
            break
    return DataFrame

Archivos = !ls precios/*.xls*
DataFrames = []
for filename in Archivos:
    DataFrames.append(Lectura(filename))
DataFrames = pd.concat(DataFrames)
DataFrames = pd.melt(DataFrames,id_vars = 'Fecha')
DataFrames.columns = ['Fecha','Hora','Precio']
```

**2.--** Compute e imprima el número de registros con datos faltantes.

Con la función dropna, se conocen los registros que tienen al menos uno de sus campos vacíos o faltantes, y este valor se le resta al total de datos

**3.--** Compute e imprima el número de registros duplicados.

```python
DataFrameSUnicos = DataFrames.copy()
DataFrameSUnicos = DataFrameSUnicos.drop_duplicates()
Duplicados =  len(DataFrames.index) - len(DataFrameSUnicos.index)
Duplicados
```

**4.--** Elimine los registros con datos duplicados o datos faltantes, e imprima la cantidad de registros que quedan (registros completos).

```python
DataFrameSUnicos = DataFrames.copy()
DataFrameSUnicos = (DataFrameSUnicos.drop_duplicates())
DataFrameSUnicos = DataFrameSUnicos.dropna()
len(DataFrameSUnicos.index)
```

**5.--** Compute y grafique el precio primedio diario.

```python
df = DataFrames.copy()
df['dia'] = pd.DatetimeIndex(df['Fecha']).day
df['mes'] = pd.DatetimeIndex(df['Fecha']).month
aux = df.groupby('dia')['Precio'].mean()
lines = aux.plot.line(color="purple", title="Precio promedio diario", linewidth=4, figsize=(10,4))
lines.set_xlabel("Día", fontsize=10)
lines.set_ylabel("Precio promedio", fontsize=10)
```

**6.--** Compute y grafique el precio máximo por mes.

```python
aux = df.groupby('mes')['Precio'].max()
lines = aux.plot.line(color="red", title="Precio máximo por mes", linewidth=4, figsize=(10,4))
lines.set_xlabel("Mes", fontsize=10)
lines.set_ylabel("Precio máximo", fontsize=10)
```

**7.--** Compute y grafique el precio mínimo mensual.

```python
aux = df.groupby('mes')['Precio'].min()
lines = aux.plot.line(color="green", title="Precio mínimo por mes", linewidth=4, figsize=(10,4))
lines.set_xlabel("Mes", fontsize=10)
lines.set_ylabel("Precio mínimo", fontsize=10)
```

**8.--** Haga un gráfico para comparar el precio máximo del mes (para cada mes) y el precio promedio mensual.

Nota: Las barras azules representan el precio máximo por mes y las naranjas, el precio promedio por mes

```python
aux = df.groupby('mes')['Precio'].max()
aux2 = df.groupby('mes')['Precio'].mean()
prueba = pd.concat([aux, aux2], axis=1)
lines = prueba.plot(kind="bar", title= "Precio máximo vs. Precio promedio", figsize=(10,4))
lines.set_xlabel("Mes", fontsize=10)
lines.set_ylabel("Precio", fontsize=10)
```
